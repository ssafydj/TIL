ClassMate 클래스는 학생 인스턴스 2개를 인자로 받아서 인스턴스를 만든다.(한 커플이 만들어진 것)

`sum_age()` 메서드는 커플의 나이를 합해서 출력한다.

`is_teen()` 메서드는 두 커플이 모두 10대일 시 True 를 출력한다.
