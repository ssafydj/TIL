# `all()`

> all은 인자로 받는 iterable(range, list)의 모든 요소가 참이거나 비어있으면 True를 반환합니다.(대우: 하나라도 False면 False!)
> 문제의 대우를 생각해보면 쉽게 풀린다.
> 이와 같은 my_all(x)을 작성해보세요

```python
# 예제 입력 및 출력
print(all([1, 2, 5, '6']))
print(all([[], 2, 5, '6']))
```

```python
# 아래에 코드를 작성해주세요.

def my_all(elements):
    for element in elements:
        if not element:                 # == if bool(element) == False
            return False
    else:
        return True                             # True False Complex 등 예약어는 ''없이 그냥 사용.
        
```

```python
# 예제 입력 및 출력
print(my_all([1, 2, 5, '6']))
print(my_all([[], 2, 5, '6']))
```

# `any()`

> any는 인자로 받는 iterable(range, list)의 요소 중 하나라도 참이면 True를 반환하고, 비어있으면 False를 반환합니다.
>
> 이와 같은 my_any(x)를 작성해보세요.

```python
# 예제 입력 및 출력
print(any([1, 2, 5, '6']))
print(any([[], 2, 5, '6']))
print(any([0]))
```

```python
# 아래에 코드를 작성해주세요.
# 하나의 값만 True 면 True! + 빈 값은 False!

def my_any(x):
    for i in x:
        if bool(i) == True:
            return True
                
    else:
        return False
```

```python
def my_any(x):
    for i in x:
        if i:
            return True
    return False
```

```python
# my_any 예제 입력 및 출력
print(my_any([1, 2, 5, '6']))
print(my_any([[], 2, 5, '6']))
print(my_any([0]))

print(bool([0]))
```
